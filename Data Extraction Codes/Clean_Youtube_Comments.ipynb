```python
def clean_text(text):
    text = re.sub('<.*?>', '', text)
    text = re.sub('[^a-zA-Z]', ' ', text).lower()
    words = nltk.word_tokenize(text)
    words = [w for w in words if w not in stopwords.words('english')]
    stemmer = PorterStemmer()
    words = [stemmer.stem(w) for w in words]
    text = ' '.join(words)
    return text
```

```python
for i in range(1,17):
  print(i)
  movie_comments = pd.read_csv("/content/movieComments/movieComments"+str(i)+".txt",delimiter = "delimeterByIshita")
  movie_comments = movie_comments.dropna()
  movie_comments['cleaned_comments'] = movie_comments['comments'].progress_apply(clean_text)
  movie_comments.to_csv("movieCommentsCleaned/movieCommentsCleaned"+str(i)+".csv")
  
```
