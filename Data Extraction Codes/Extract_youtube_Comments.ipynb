```python
def get_comments(id):
  comments = video_comments(id)
  print(id, end = " ")

  if(type(comments) != list):
    print("w", end=" ")

  else:
    print(".",end = " ")
    if(len(comments)>=100):
      comments = comments[-100:]
    comments = [id+'delimeterByIshita' + s for s in comments]
    with open("movieComments16.txt", "a", encoding='utf-8') as f:
      for i in comments:
        f.write(i+ '\n')
```
